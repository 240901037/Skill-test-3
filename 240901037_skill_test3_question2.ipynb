```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
# Features: [Voltage, Current]
X = [
    [5,0.5],[5.5,0.4],[4.8,0.6],
    [6,0.3],[6.5,0.2],[6.2,0.25],
    [4.9,0.55],[5.8,0.35],[5.2,0.5],[6.3,0.22]
]
# Labels: Working = 0, Defective = 1
y = [0,0,0,1,1,1,0,1,0,1]
# Create and train KNN model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
# New component
new_data = [[5.4, 0.45]]
# Prediction
result = model.predict(new_data)
# Output
print("Component Status:", "Working" if result[0]==0 else "Defective")
# Visualization (Voltage vs Current)
voltage = [i[0] for i in X]
current = [i[1] for i in X]
plt.scatter(voltage, current, c=y)
plt.scatter(new_data[0][0], new_data[0][1], c='red', marker='x')
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("KNN Component Classification")
plt.show()
```
